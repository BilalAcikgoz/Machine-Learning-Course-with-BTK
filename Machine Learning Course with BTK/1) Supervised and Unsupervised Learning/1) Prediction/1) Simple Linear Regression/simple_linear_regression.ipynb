```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

veriler = pd.read_csv('satislar.csv')
veriler
```

```python
# data preprocessing
aylar = veriler[['Aylar']]
print(aylar)

satislar = veriler[['Satislar']]
print(satislar)
```

```python
#verilerin egitim ve test icin bolunmesi
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(aylar,satislar,test_size=0.33, random_state=0)

#verilerin olceklenmesi
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
Y_train = sc.fit_transform(y_train)
Y_test = sc.fit_transform(y_test)
```

```python
# model inşası
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train) # Scale olmadan eğitim yaptırdık
```

```python
tahmin = lr.predict(x_test)
print(tahmin) # tahminler büyük çoğunlukla tutuyor
print(y_test) # y_test'in amacı tahminleri karşılaştırmak
```

```python
# veri görselleştirme

x_train = x_train.sort_index() # train-test split yaparken rastgele split yapıldığı için veriler sıralı değil
y_train = y_train.sort_index()

plt.plot(x_train, y_train)
plt.plot(x_test, tahmin)

plt.title("Aylara Göre Satış")
plt.xlabel("Aylar")
plt.ylabel("Satışlar")
```
